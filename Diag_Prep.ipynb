```python
#The diagnosis file was treated initially individually as it had a very large number of columns
import pandas as pd
#Imported the Excel file Diagnosis_df
diag_df = pd.read_excel('Diagnosis_file.xlsx', engine='openpyxl')

# Display the first few rows
print(diag_df.head())
```

```python
#the column type is converted to string 
diag_df['DIAGNOSIS'] = diag_df['DIAGNOSIS'].fillna('').astype(str)

#Values are split by ','

diag_df['DIAGNOSIS'] = diag_df['DIAGNOSIS'].str.lower() #values all in lower case 
diag_df['DIAGNOSIS'] = diag_df['DIAGNOSIS'].str.split(' ; ')

print(diag_df.head())
```

```python
# Values exploded into separate rows
diag_df = diag_df.explode('DIAGNOSIS')
#Drop duplicates
diag_df_clean = diag_df.drop_duplicates()

#Elimintating redundant values that have white spaces in the end and the beginning of the value 
diag_df_clean2 = diag_df_clean.copy()
diag_df_clean2['DIAGNOSIS'] = diag_df_clean2['DIAGNOSIS'].astype(str)
diag_df_clean2['DIAGNOSIS'] = diag_df_clean2['DIAGNOSIS'].str.strip()
diag_df_clean2 = diag_df_clean2.drop_duplicates(subset=['DIAGNOSIS'])

print(diag_df_clean2)
```

```python
diag_df_clean2 = diag_df_clean2.sort_values(by='DIAGNOSIS')
#Transpose the dataframe to make the rows columns 
diag_df_clean_T = diag_df_clean2.T
print("\nTransposed DataFrame:", diag_df_clean_T)
```

```python
diag_df_clean_T.columns = diag_df_clean_T.iloc[0]   # set first row as header
# drop the first row since it's now the header
diag_df_clean_T = diag_df_clean_T[1:]               
# Reset index 
diag_df_clean_T = diag_df_clean_T.reset_index(drop=True)
print(diag_df_clean_T)
```

```python
from collections import Counter

words = []
for col in diag_df_clean_T.columns:
    words.extend(col.split(" "))   
# Count frequency of words
word_counts = Counter(words)

# Find the most common word
most_common_word, freq = word_counts.most_common(1)[0]

print("Most redundant word:", most_common_word)
print("Frequency:", freq)
print(word_counts)
```

```python
#To minimize the columns and overcome the non-standardised entries, columns are clustered and deleted one group of clusters, by one and checked every time for the number of columns remaining to work with 
#Droping columns that contain 'pregnancy', in order to cluster them in one column
for col in diag_df_clean_T.columns:
    if "pregnancy" in col.lower():   # case-insensitive check
        diag_df_clean_T = diag_df_clean_T.drop(columns=[col])

# create an empty column for clustering
diag_df_clean_T["pregnancy"] = None   

print(diag_df_clean_T.shape)
```
